```python
#Scatter Plot
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance = pd.read_csv('mydata/insurance.csv')
insurance.head(5)
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses)
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, color = '#91bd3a')
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, color = '#339900', marker = '>')
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.sex)
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.smoker)
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.xlim([0,55])
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.smoker)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.gcf().text(.5, .9, "Scatter Plot", fontsize = 40, color = 'black', ha = 'center', va = 'center')
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.smoker, style = insurance.sex)
plt.show()
```

```python
stdperf = pd.read_csv('mydata/StudentsPerformance.csv')
stdperf.head()
```

```python
plt.figure(figsize = (9,9))
sns.scatterplot(x = stdperf['reading score'], y = stdperf['writing score'],
               hue = stdperf['test preparation course'], size = stdperf['math score'])
plt.show()
```

```python
#Relplot -> Line & Scatter 
```

```python
sns.set_style('white')
```

```python
employment = pd.read_csv('mydata/employmentx.csv', delimiter = ';')
employment.head()
```

```python
plt.figure(figsize = (14,16))
sns.relplot(x = 'Period', y = 'Unemployed', hue = 'Gender', data = employment, height = 7)
plt.show()
```

```python
plt.figure(figsize = (14,16))
sns.relplot(x = 'bmi', y = 'expenses', hue = 'smoker', data = insurance, height = 8)
plt.show()
```

```python
sns.set(rc = {'xtick.labelsize' : 12, 'ytick.labelsize' : 12, 'axes.labelsize' : 12, 'axes.grid' : False})
sns.relplot(x = 'bmi', y = 'expenses', hue = 'smoker', data = insurance, height = 8, 
           palette = ['#FF3333', '#00CC00'])
plt.show()
```

```python
fish = pd.read_csv('mydata/Fish.csv')
fish.head()
```

```python
plt.figure(figsize = (14,16))
sns.relplot(x = 'Width', y = 'Height', hue = 'Species', data = fish, height = 8)
plt.gcf().text(.5, .99, "Rel Plot", fontsize = 40, color = 'black', ha = 'center', va = 'center')
plt.show()
```

```python
sns.set(rc = {'xtick.labelsize' : 12, 'ytick.labelsize' : 12, 'axes.labelsize' : 12, 'axes.grid' : False})
sns.relplot(x = 'bmi', y = 'expenses', hue = 'sex', col = 'sex', row = 'region', data = insurance)
```

```python
#Bar Plot
```

```python
pokemon = pd.read_csv('mydata/pokemon.csv')
pokemon.head(20)
```

```python
plt.figure(figsize = (20,7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Attack'])
plt.show()
```

```python
plt.figure(figsize = (20,7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Defense'])
plt.show()
```

```python
plt.figure(figsize = (20,7))
order = pokemon.groupby(['Type 1']).mean().sort_values('Defense', ascending = False).index
sns.barplot(x = pokemon['Type 1'], y = pokemon['Defense'], order = order)
plt.show()
```

```python
plt.figure(figsize = (20,7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Speed'])
plt.show()
```

```python
plt.figure(figsize = (8, 10))
sns.barplot(x = pokemon['Speed'], y = pokemon['Type 1'])
plt.show()
```

```python
#Countplot
```

```python
plt.figure(figsize = (20, 7))
sns.countplot(x = pokemon['Type 1'])
plt.show()
```

```python
plt.figure(figsize = (20,7))
sns.countplot(x = pokemon['Type 1'], order = pokemon['Type 1'].value_counts().index)
plt.show()
```

```python
exercise = pd.read_csv('mydata/exercise.csv')
exercise.head()
```

```python
exercise = exercise.drop('Unnamed: 0', axis = 1)
exercise.head()
```

```python
plt.figure(figsize = (8,6))
sns.barplot(x = exercise.kind, y = exercise.pulse, hue = exercise.diet)
plt.show()
```

```python
#Cat Plot
```

```python
sns.catplot(y = 'Type 1', kind = 'count', data = pokemon)
plt.show()
```

```python
sns.catplot(y = 'Type 1', kind = 'count', data = pokemon, order = pokemon['Type 1'].value_counts().index)
plt.show()
```
