```python
#Pair Plot
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fish = pd.read_csv('mydata/Fish.csv')
fish.head()
```

```python
fish1 = fish[fish['Species'].isin(['Bream', 'Perch', 'Pike'])]
fish1.head()
```

```python
plt.figure(figsize = (11,9))
sns.pairplot(fish1, hue = 'Species')
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', palette = 'husl', height = 2)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Length1", "Weight", "Height", "Width"], height = 2)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Length1", "Weight", "Height", "Width"], 
            markers = ['o', 'D', '<'], height = 3)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], 
            markers = ['o', 'D', '<'], height = 5)
plt.show()
```

```python
#Lineer Regresyon modelini scatter plot üzerine uygulayabiliriz.
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], kind = 'reg', height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], plot_kws = dict(s = 140),
             height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], plot_kws = dict(s = 140, alpha = .6),
             height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], 
             plot_kws = dict(s = 140, linewidth = 1, alpha = 0.6), height = 5)
plt.show()
```

```python
#Pair Grid
```

```python
g = sns.PairGrid(fish1)
g = g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(fish1)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype = "step", linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
#Heatmap
```

```python
canada = pd.read_csv('mydata/canada1.csv', delimiter = ';')
canada.head()
```

```python
canada = canada.drop(columns = ['AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName', 'Type', 'Coverage'])
canada.head()
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
```

```python
canada.head()
```

```python
canada1 = canada.loc['2004' : , ['Germany', 'Afghanistan', 'Australia', 'El Salvador', 'Chile', 'Paraguay', 'Austria', 'Brazil', 'Uruguay', 'Jamaica']]
canada1.head()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, center = 4000)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'coolwarm')
plt.yticks(rotation = 0)
plt.show()
```

```python
#Aradaki hücreleri daha belirgin hale getirme
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, linewidths = .1)
plt.yticks(rotation = 0)
plt.show()
```

```python
#Kutucukların içine sayı yerleştirme
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, annot = True, fmt = 'd')
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'YlGnBu', annot = True, fmt = '.2f')
plt.yticks(rotation = 0)
plt.show()
```
