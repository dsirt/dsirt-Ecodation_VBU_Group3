```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1" : col1, "C2" : col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid":True, 'xtick.labelsize' : 14, 
              'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid":False, 'xtick.labelsize' : 14, 
              'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, color = '#FF5722', linewidth = 2)
sns.lineplot(x = df.C1, y = np.cos(df.C1), data = df, color = '#FFEB3B', linewidth = 2)
plt.show()
```

```python
canada = pd.read_csv("mydata/canada1.csv", delimiter = ';')
canada.head()
```

```python
canada.columns
```

```python
canada = canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'])
canada.head()
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
```

```python
canada.head()
```

```python
plt.figure(figsize = (14,6))
plt.title('Immigrants from Turkey', fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid":False, 'xtick.labelsize' : 14, 
              'ytick.labelsize' : 14})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'])
plt.show()
```

```python
plt.figure(figsize = (14,6))
plt.title('Immigrants from Turkey', fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid":False, 'xtick.labelsize' : 14, 
              'ytick.labelsize' : 14})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'], color = '#ff9900', label = 'Turkey')
sns.lineplot(x = canada.index.values, y = canada['Italy'], color = '#4586ff', label = 'Italy')
sns.lineplot(x = canada.index.values, y = canada['Indonesia'], color = '#a2ef44', label = 'Indonesia')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```
